## Part 1. Variable Types and Levels of Measurement
In statistics, a variable has two defining characteristics:
 - A variable is an attribute that describes a person, place, thing, or idea.
 - The value of the variable can "vary" from one entity to another.

<a id="variables"></a>
### 1.1 Dependent and Independent Variables
An **Independent Variable**, also called a predictor variable, is a variable that is expected to have an effect on another **dependent variable**, sometimes called an outcome variable. Thus a dependent variable is simply that, a variable that is dependent, at least in part, on an independent variable.

<a id="research"></a>
### 1.2 Experimental and Non-Experimental Research
The degree to which an analyst can influence independent and dependent variables in their research depends on their research design.

 - **Experimental research:** In experimental research, the aim is to manipulate one or more independent variables and then examine the effect on one or more dependent variables. Since it is possible to manipulate the independent variable(s), experimental research, especially a [randomized controlled trial](https://en.wikipedia.org/wiki/Randomized_controlled_trial), has the advantage of establishing stronger causal links between variables. 
 
- **Non-experimental research:** Non-experimental research lacks the manipulation of an independent variable, random assignment of participants to conditions or orders of conditions, or both. These conditions often occur because it is either impractical or unethical manipulate the independent variables or use random assignment. For example, a researcher may be interested in the effect of illegal, recreational drug use (the independent variable(s)) on certain types of behaviour (the dependent variable(s)). It would be unethical for the researched to forcibly subject research participants to drug use. 

<a id="cat_cont"></a>
### 1.3 Categorical, Continuous and Discrete Variables
A **categorical variable** is comprised of observations that can take on one of a limited, and usually fixed, number of possible values based on some qualitative property. Categorical variables are sometimes called *qualitative* or *discrete* variables.
 - Example: Gender

A **continuous variable**, also known as a quantitative variable, has an infinite number of possible values and therefore an infinite number of possible values between any two values. In other words, any value is possible for the variable. A continuous variable is the opposite of a categorical variable, which can only take on a certain number of values.
 - Example: Height

A **discrete variable** can only take on a finite number of values. All categorical (i.e. qualitative) variables are discrete while only some quantitative variables are discrete. An example of a discrete quantitative variable would be a performance rating on a 1-5 scale or temperature rounded to the nearest degree. Although somewhat [controversial](https://link.springer.com/article/10.1007/BF02595397), a variable that takes on enough discrete values could be considered continuous for practical purposes. One example is time to the nearest millisecond. 
 - Example: Number of customer complaints

<a id="levels"></a>
### 1.4 Levels of Measurement
**Level of measurement** refers to the relationship among the values that are assigned to the attributes of a variable. An example will elucidate this relationship:

<img src="assets/variables.png" width = 400 height = 400>

Begin with the idea of the variable, in this example "party affiliation." 

That variable has a number of attributes:  republican, democrat, and independent. 

We also arbitrarily assign the values 1, 2, and 3 to these three attributes. 

The **level of measurement** describes the relationship among these three values.

#### Levels of Measurement for Categorical Variables

- **Nominal** - variables that have two or more categories, but which do not have an intrinsic order.
    - Values within a nominal measure are:
        - all-inclusive
        - mutually exclusive
    - Examples:
        - Nationality
        - Social Security Number
    - Takeaway:
         - When a nominal variable has only two categories, it is a **discrete dichotomous variable**. For example: dead or alive.

 
- **Ordinal** - nominal variables but with categories that can be ordered or ranked.
    - Values within an ordinal measure are:
        - all-inclusive
        - mutually exclusive
        - ordered, but the "distance" between two adjacent categories may vary
    - Examples:
        - Airline ticket class
        - Likert scale (e.g. 1 - strongly disagree, 2 - disagree, 3 - neutral, 4 - agree, 5 - strongly agree)
    - Takeway:
        - Since the distance between two categories can vary, there's disagreement on whether to analyze ordinal data [parametrically](https://en.wikipedia.org/wiki/Parametric_statistics) or [non-parametrically](https://en.wikipedia.org/wiki/Nonparametric_statistics).
        - When an ordinal variable has only two categories, it is a **continuous dichotomous variable**. A continuous dichotomous variable might also be based on an underlying measurement. For example, young or old.

#### Levels of Measurement for Continuous and Discrete Quantitative Variables
- **Interval** - variables that can be measured along a continuum with a numerical value
    - Values within an interval measure are:
        - all-inclusive
        - mutually exclusive
        - ordered
        - equidistant from adjacent elements
        - arbitray when equal to zero
    - Examples:
        - Farenheight or Celsius
        - Time
    - Takeaway:
        - Without a true zero, it is nonsensical to compute ratios. For example, there is no such thing as “no temperature.”  Without a true zero, there is no sense in which the ratio of 40 to 20 degrees Fahrenheit is the same as the ratio of 100 to 50 degrees. After all, if we were to change the "zero" Fahrenheit label to 10 degrees, the two ratios would instead be 30 to 10 and 90 to 40, which is no longer the same! 


- **Ratio** - interval variables, but with the added condition that 0 (zero) indicates and absence of that variable
     - Values within a ratio measure are:
        - all-inclusive
        - mutually exclusive
        - ordered
        - equidistant from adjacent elements
        - meaningful when equal to zero
     - Examples:
        - Car prices
        - Weight
     - Takeaway:
        - The term "ratio" reflects the fact that you can use the ratio of measurements. This is because a value of zero means complete absence of the variable. So, for example, a distance of 20 miles is twice the distance of 10 miles.

#### Levels of Measurement - A Hierarchy
It's important to know that there is a hierarchy among the levels of measurement. At lower levels, you're making fewer assumptions about your data and so your analyses *tend* to be less sensitive. As you move up the hierarchy, you expand your list of assumptions. In general, higher levels of measurement (e.g. interval or ratio) yield more enlightening analyses.

#### Why are Levels of Measurement Important?
Knowing your data's levels of measurement helps you:
 - Interpret the data, especially if categorical variables have been recoded to numerics for data compression
 - Decide which statistical analyses are appropriate.
 - Communicate your assumptions to stakeholders.

### 1.6 Measurement Theory Deep Dive:  Miles per Gallon or Gallons per Mile?
The conclusions of a statistical analysis shouldn't depend on an arbitrary measurement decision. If we want our analyses to say something about reality, we need to be sure that such arbitrary decisions are removed from the equation.

Let's consider an analysis of fuel efficiency. 

If we were just interested in the average distance that can be driven with a given amount of gas, we would logically speak in terms of miles per gallon. 

If we were interested in the average amount of gas required to drive a given distance, we would rather speak in terms of gallons per mile. 

Although both miles per gallon and gallons per mile are measurements of fuel efficiency, they can yield different results when they're the inputs for a statistical test. Let's see this in action with some dummy data:

```python
gpm = [1/40, 2/40, 3/40, 4/40]
mpg = [40/1, 40/2, 40/3, 40/4] # reciprocals of the gpm list
vehicle_weight = [100, 150, 200, 250] # we'll add in a third variable so we can run a simple statistical test
dummy_data = pd.DataFrame({'gallons per mile': gpm, 'vehicle weight': vehicle_weight, 'miles per gallon':mpg})
dummy_data
```

```python
from scipy.stats.stats import pearsonr
pearsonr(dummy_data['vehicle weight'],
         dummy_data['gallons per mile']) #Calculate a Pearson correlation coefficient and the p-value
```

```python
pearsonr(dummy_data['vehicle weight'],
         dummy_data['miles per gallon']) #Calculate a Pearson correlation coefficient and the p-value
```

As you can see, correlating gallons per mile with vehicle weight gives a large negative correlation while correlating miles per gallon gives almost the exact opposite. Also note the differing p values.

Our challenge is to make inferences about fuel efficiency that do not depend on our measurement choice. The trick is to *transform* our fuel efficiency measures in a way that preserves the relationships between observations.

We can do that by recognizing that both miles per gallon and gallons per mile are measurements of the same attribute on a log-interval scale. This means that the logarithm of either can be treated as a measurement on an interval scale.

Naturally there's math involved in this, so consider the following identity:

$$log(a/b) = log(a) - log(b)$$



```python
# apply log10 to each observation in the gallons per mile and miles per gallon columns and create two new columns
dummy_data['log_gpm'] = dummy_data['gallons per mile'].map(lambda x: np.log10(x))
dummy_data['log_mpg'] = dummy_data['miles per gallon'].map(lambda x: np.log10(x))
dummy_data
```

Notice now how our log transformed measures are identical except for their sign. Let's see what happens when we recalculate Pearson's R:

```python
print(pearsonr(dummy_data['vehicle weight'],
               dummy_data['log_gpm']))

print(pearsonr(dummy_data['vehicle weight'],
               dummy_data['log_mpg']))
```

Now you can see that we get the same p values and correlation coefficients (except the sign). In your free time, think about why the correlation coeffecients have different signs.

#### Deep Dive Takeaway
Measurement theory shows that strong assumptions are required for certain statistics to provide meaningful information about reality. You should always think about the meaning of your data and question the assumptions behind your analysis (ideally before somebody else does).

### 2.3 The Mean
The sum of *n* numbers divided by *n*. Mathematically,

$$ ( \frac{1}{n} \sum_{i=i}^{n} x_{i} ) = { {x^1 + x^2 +...+x^n} \over{n} }$$

### 2.4 The Median
To find the median, the data should first be arranged in order from least to greatest. If there is an even number of values in the data set, then the median is found by taking the mean (average) of the two middlemost numbers. Otherwise the median is the value separating the higher half of a dataset from the lower half.

```python
# What is the median mpg?

# One way (using the built-in pandas method)
data.mpg.median()

# Another way (using a function)
def getMedian(lst):
    n = len(lst)
    if n < 1:
        return None
    if n % 2 == 1:
        return sorted(lst)[n//2]
    else:
        return sum(sorted(lst)[n//2-1:n//2+1])/2.0

getMedian(data.mpg)

# Yet another way (using the median method from the numpy module)
np.median(data.mpg)
```

## Part 3. Crosstabs and Distribution Shapes
Counting the frequency of occurences within each category of a particular categorical variable can also describe your dataset. These counts can be organized in tables known as frequency tables or crosstabs.

Crosstabs can present the frequencies for either a single or multiple categorical variables.

Crosstabs can also display relative frequencies by showing the number of times a certain event occurs in relation to the overall population.

<a id="oneWay"></a>
### 3.1 One-Way Tables
A one-way table will show the frequency distribution for a single categorical variable. 

Let's make a crosstab that counts the number of cars of each cylinder type:

```python
cyl_tab = pd.crosstab(index = data['cylinders'], columns = 'count')
cyl_tab
```

Even a simple one-way table like the one above can give us some useful insight: we immediately see that there are very few vehicles with 3 or 5 cylinders.

Since the crosstab function produces a pandas DataFrame, all of the common DataFrame operations will work on it. For example, say we want to know the relative frequency instead of the counts for each cyliner:

```python
cyl_tab / cyl_tab.sum()
```

The count column now has the proportion of observations of each cylinder type.

<a id="twoWay"></a>
### 3.2 Two-Way Tables
Two-way crosstabs, also called contingency tables, show counts across two dimensions. This allows you to get an idea for the relationship between two categorical variables. To create a two way table, you just pass two column names to the pd.crosstab() function:

```python
cyl_origin_tab = pd.crosstab(index = data['cylinders'], columns = data['origin'])
cyl_origin_tab.columns = ["American","European","Japanese"] #renaming the columns
cyl_origin_tab
```

You can display the marginal counts (row and column totals) by including the argument margins=True:

```python
cyl_origin_tab_margin = pd.crosstab(index = data['cylinders'], columns = data['origin'], margins = True)
cyl_origin_tab_margin.columns = ["American","European","Japanese","All"]
cyl_origin_tab_margin
```

And just like with one-way crosstabs, you can use DataFrame operators to get the relative frequencies for each cell:

```python
cyl_origin_tab_margin / cyl_origin_tab_margin.ix['All','All'] #divide every cell value by the total # of observations (397)
```

Of course it might be more useful to get the proportion of counts along each column (in this case, the cylinder proportion for each origin). To do that, divide by the column totals:

```python
cyl_origin_tab_margin/ cyl_origin_tab_margin.ix['All']
```

You can also get the proportion of counts along each row (in this case, the origin proportion for each cylinder) by dividing each cell by the row totals:

```python
cyl_origin_tab_margin.div(cyl_origin_tab_margin["All"], #divide by the 'All' column
                        axis=0) #axis = 0 to indicate that we want to go down the columns row-by-row
```

<a id="high"></a>

## Part 4. Variance and Standard Deviation
Variance and standard deviation are two more descriptive statistics that will help you get a better idea of how concentrated your data is around the mean. 

**Variance:** The average of the squared differences of each observation from the mean. Mathematically,

$$\sigma^2 = \frac{\sum_{i=1}^N (X_i - \mu)^2}{N}$$

> where *N* is the number of samples, $X_i$ is each observation within the population, and $\mu$ is the mean of that population.


**Standard Deviation:** The square root of the variance. Mathematically,

$$\sigma = \sqrt{\sigma^2}$$

*So why do we bother taking the square root of the variance to calculate standard deviation?*

Since variance involves squaring differences, the variance does not have the same unit of measurement as the original observations. For example, distances measured in miles ($m$) have a variance measured in miles squared ($m^2$). So if we take the square root of the variance we'll get a number in the original units. That's the standard deviation! You can also standardize the standard deviation by calculating the [Coefficient of Variation](https://en.wikipedia.org/wiki/Coefficient_of_variation)!

Let's calculate the variance of mpg in our cars dataset.

To do this, we need to know:
 - The number of observations ( $N$ ) 
 - The mean mpg ( $\mu$ ):

```python
mean_mpg = data['mpg'].mean()
N = data['mpg'].count()
print("The mean mpg is",mean_mpg, "and the number of observations is", N)
```

Next we need to calculate the squared distance between each observation $X_i$ and the mean $\mu$. Then we'll need to sum all of these differences. We can do that with the following code:

```python
sum_of_differences = sum((data['mpg'] - mean_mpg)**2)
print("The sum of differences is {}".format(sum_of_differences))
```

Finally, we divide the sum of differences by N:

```python
variance = sum_of_differences / N
print('The variance for mpg is', variance)
```

Phew! That took a lot of work and a lot of steps. If only there was a way to do that more quickly...

```python
data.mpg.var(ddof = 0) #we'll explore in a later lesson why we set ddof to 0
```

```python
# Calculating the standard deviation:
data.mpg.std(ddof = 0)
```

### Interpreting Standard Deviation
Standard deviation can be difficult to interpret on its own. Basically, a small standard deviation means that the values in a dataset are close to the mean while a large standard deviation means that the values are farther away from the mean. This is often easier to determine visually, which is the subject of the next section.

However, the standard deviation is a very important measure when it comes to inferential and predictive statistics. So you should become familiar with it!

## Part 5. Shapes of Distributions
To get a better feel for your data's variability, you can graph how your data is distributed across its range. This will give you a better feel for the data's *symmetry* and *skew*.

But before we visually explore the distributions in our car dataset, we need to create, and acquaint ourselves with, the ideal types.

```python
# This code will create some dummy data to help us visualize the ways in which we describe the shape of a distribution

norm_data = np.random.normal(size=100000) #create a normal (Gaussian) distribution

skewed_data = np.concatenate((np.random.normal(size=35000)+2, 
                             np.random.exponential(size=65000)), 
                             axis=0) # create skewed distribution

uniform_data = np.random.uniform(-1,1, size=100000) #create uniform distribution 

tailed_data = np.concatenate((np.random.exponential(size=50000),
                             np.random.exponential(size=50000)*(-1)),
                             axis=0) #created tailed distribution

data_df = pd.DataFrame({"norm":norm_data,
                       "skewed":skewed_data,
                       "uniform":uniform_data,
                       "tailed":tailed_data}) #put each distribution into a dataframe

data_df.head() #print top 5 rows so you can inspect the data frame we just built
```

<a id='normal'></a>
### 5.1 The Normal Distribution
In probability theory, the normal (or Gaussian) distribution is a very common continuous probability distribution. Normal distributions are important in statistics and are often used in the natural and social sciences to represent real-valued random variables whose distributions are not known. Mathematically,

$$f(x)=\dfrac{1}{\sigma \sqrt{2\pi}} \text{exp}\left\{-\dfrac{1}{2} \left(\dfrac{x-\mu}{\sigma}\right)^2\right\}$$

> when the mean $\mu$ equals 0 and the standard deviation $\sigma$ equals 1

For now it's not important to understand the math. Just know, however, that the intuition behind this math is what will allow us to make the leap from descriptive to parametric statistics in the next lesson.



For now, let's just focus on the characteristics of a normal distribution.

<img src='assets/normalDist.png' width = 400 height = 400 >

A normal distribution is a bell-shaped frequency distribution curve. The bell shape means that most of the data clusters around the mean ( $\mu$). Since this is a frequency distribution, the further a data point is from the mean, the less likely it is to occur. This is related to the concept of outliers.

There are many things, such as height, intelligence, and the size of objects produced by machines that naturally *approximate* a normal distribution.

#### Characteristics of Normal Distribution

1. symmetric - the left half is a mirror image of the right half
2. unimodal - there is only one mode
3. asymptotic - the tails approach but never touch the x-axis
4. the mean, median, and mode are all equal.

These characteristics mean that:
 - approximately 99% of the data is within 3 standard deviations of the mean.
 - approximately 95% of the data is within 2 standard deviations of the mean.
 - approximately 68% of the data is within 1 standard devation of the mean.

#### Fun Fact
Even if we don't expect a normal distribution in our population, [Chebyshev's inequality](https://en.wikipedia.org/wiki/Chebyshev%27s_inequality) tells us that (usually) at least 75% of values lie within two standard deviations of the mean and 89% within three standard deviations.

<a id="peaked"></a>
### 5.2 Kurtosis
Kurtosis occurs when you have a heavily-tailed distribution. These distributions will often look like they have sharp peaks, but high kurtosis is actually the result of infrequent extreme deviations (or outliers), as opposed to frequent modestly sized deviations.

Here's what it looks like in comparison to the normal distribution:

```python
data_df[["tailed",'norm']].plot(kind="density", xlim=(-4,4))
```

#### Measuring Kurtosis
You can measure kurtosis!

$$\mbox{kurtosis} = \frac{\sum_{i=1}^{N}(Y_{i} - \bar{Y})^{4}/N} {s^{4}}  - 3$$

> where $\bar{Y}$ is the mean, $s$ is the standard deviation, and $N$ is the number of data points.

High kurtosis will approach a value of 3 while a normal distribution has approximately 0 kurtosis. Thus positive kurtosis indicates a "heavy-tailed" distribution while negative kurtosis indicates a "light-tailed" distribution.

```python
tail_kurt = data_df['tailed'].kurt()
norm_kurt = data_df['norm'].kurt()

print("Normal data has kurtosis of {} while the tailed data has kurtosis of {}".format(norm_kurt, tail_kurt))
```

<a id="skew"></a>
### 5.3 Skew
Skewness is a measure of the asymmetry of a probability distribution function. Skew can be positive, negative, or undefined.

Let's see negative (aka left) skew in comparison to a normal distribution

```python
data_df[["norm","skewed"]].plot(kind="density", xlim=(-5,5))
```

When there is left/negative skew:
 - the mean is typically less than the median
 - the tail of the distribution is longer on the left hand side than on the right hand side
 - the median is closer to the third quartile than to the first quartile
 
The opposite is true for right/positive skew

#### Measuring Skew
Here's the formula for skew:

$$g_{1} = \frac{\sum_{i=1}^{N}(Y_{i} - \bar{Y})^{3}/N} {s^{3}}$$

> where $\bar{Y}$ is the mean, $s$ is the standard deviation, and $N$ is the number of data points.

```python
tail_skew = data_df['skewed'].skew()
norm_skew = data_df['norm'].skew()

print("Normal data has skew of {} while the skewed data has skew of {}".format(norm_skew, tail_skew))
```

The skew for a perfectly normal distribution is zero. Thus highly symmetric data has skew near zero.

Negative values for the skew indicate data that are skewed left and positive values for the skewness indicate data that are skewed right.

### Part 6. Covariance and Correlation
As the co- prefix in these terms implies, covariance and correlation help us understand the relationship between the distributions of *two* variables. 

More specifically, they describe the degree to which random variables tend to jointly deviate from their expected values.

<a id="cov"></a>
### 6.1 Covariance
Covariance is a measure of the extent to which corresponding elements from two sets of ordered data move in the same direction. 

The covariance between two jointly distributed real-valued random variables $X$ and $Y$ is defined as the expected product of their deviations from their individual expected values, i.e. the mean, divided by the number of observations. Mathematically,

$$Cov(X,Y)=\frac{\sum_{i=1}^{N}(X_i-\bar{X})\,(Y_i-\bar{Y})}{N}$$

#### Interpreting Covariance
 - If the greater values of one variable correspond with the greater values of the other variable, and the same holds true for the lesser values, then covariance is positive. 

 - In the opposite case - when the greater values of one variable mainly correspond to the lesser values of the other -  then the covariance is negative. 

 - The sign of the covariance therefore shows the tendency in the linear relationship between the variables. The magnitude of the covariance, however, is not easy to interpret (more on this later).

So let's calculate the covariance for two of our car variables:  mpg and weight

```python
# and let's do it by hand
mpg_mean = data['mpg'].mean()  #get the mean mpg
weigth_mean = data['weight'].mean()  #get the mean weight

cov_mpg_weight = sum((data['mpg'] - mpg_mean)*(data['weight'] - weigth_mean)) / (data['mpg'].count())
cov_mpg_weight
```

And of course there's a quicker way to calculate this. In fact, you can do it for every variable with a **covariance matrix**. In covariance matrix, the element in the $i, j$ position is the covariance between the $i^{th}$ and $j^{th}$ variables. Here's an example with three variables:
<img src = 'assets/covMatrix.png'>

```python
np.cov(data['mpg'], y = data['weight'], ddof = 0) #covariance matrix of mpg and weight
```

So what does a covariance of -5505.1278347048719 tell us? 

Well, it's negative, so we know that the greater values of weight correspond with the lower values of mpg. That makes sense since heavier cars seems like they'd have lower mpg. But what's the magnitude of this relationship?

Covariance can't tell us that...

Covariance is blind to, but influenced by, units of measurement. This means we can determine whether a relationship is positive or negative, but we can't measure the degree to which the variables move together. Moreover, we can't compare the covariance of one pair of variables to that of another. We need a standard unit of measurement to do these things. That brings us to correlation.

<a id="corr"></a>
### 6.2 Correlation
**Correlation** is a standardized measure of covariance and tells you how closely two variables move together linearly. In other words, you can use the correlation coefficient to examine the strength and direction of the linear relationship between two continuous variables.

To calculate correlation, we standardize our covariance measure by dividing by the product of the standard deviations of X and Y. Mathematically,

$$\rho_{X,Y}= \frac{\operatorname{Cov}(X,Y)}{\sigma_X \sigma_Y}$$

Let's do this manually:

```python
mpg_sd = data['mpg'].std()  # calculate standard deviation of mpg
weight_sd = data['weight'].std()  # calculate standard deviation of weight

r = cov_mpg_weight / (mpg_sd * weight_sd)  # remember we solved for covariance ealier and saved it as `covariance`
print(r)
```

### Interpreting Correlation

**Magnitude**
<br>The correlation coefficient can range in value from −1 to +1. The larger the absolute value of the coefficient, the stronger the linear relationship between the variables.


**Direction**
<br>The sign of the coefficient indicates the direction of the relationship. If both variables tend to increase or decrease together, the coefficient is positive. 

How would you interpret the correlation coefficient calculated above?

When interpreting the correlation coefficient, keep in mind the following:
 - Correlation does not mean causation! Only properly controlled experiments enable you to determine whether a relationship is causal. See these [spurious correlations](#http://www.tylervigen.com/spurious-correlations) for a good laugh.
 - The correlation coefficient is very sensitive to outliers. Usually, an outlier that is consistent with the trend of the majority of the data will inflate the correlation. An outlier that is not consistent with the rest of the data can substantially decrease the correlation.
A low Pearson correlation coefficient does not mean that no relationship exists between the variables. The variables may have a nonlinear relationship. To check for nonlinear relationships graphically, create a scatterplot or use simple regression.
 - A low correlation coefficient does not mean there's no relationship between the variables. The variables might just have a nonlinear relationship.
